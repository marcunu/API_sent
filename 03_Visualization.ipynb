```python
plt.rcParams['figure.figsize'] = (10, 7)
sns.scatterplot(x = data.SIA, y = data.Stars)
```

Apparently there is no relationship between the two variables, the only conclusion we can draw is that the chapters that scored the lowest were those whose `Sentiment Analysis` was closest to 0. In fact, the correlation between the two columns is `-0.0328`
